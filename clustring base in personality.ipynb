```python
import warnings
warnings.filterwarnings('ignore')

df.replace({'quite agree':5,'agree':4,'nothing':3,'disagree':2,'quite disagree':1},inplace=True)
df
```

```python
df=df.dropna()
df
```

```python
df.dtypes
```

```python
df['age']=df['age'].astype(int)
```

```python
df3=pd.DataFrame()
df3=df.copy()
df3
```

```python
from sklearn.preprocessing import MinMaxScaler
for col in df.columns:
    scaler=MinMaxScaler()
    df3[col]=scaler.fit_transform(df3[col].values.reshape(-1,1))
```

```python
from sklearn.cluster import KMeans
k=KMeans(n_clusters=8)
k.fit(df3)
df['cluster']=k.labels_
df
```

```python
df11=df[df['cluster']==0]
df22=df[df['cluster']==1]
df33=df[df['cluster']==2]
df44=df[df['cluster']==3]
df55=df[df['cluster']==4]
df66=df[df['cluster']==5]
df77=df[df['cluster']==6]
df88=df[df['cluster']==7]
```

```python
plt.scatter(df11.user_id,df11.cluster,color='blue')
plt.scatter(df22.user_id,df22.cluster,color='red')
plt.scatter(df33.user_id,df33.cluster,color='yellow')
plt.scatter(df44.user_id,df44.cluster,color='green')
plt.scatter(df55.user_id,df55.cluster,color='brown')
plt.scatter(df66.user_id,df66.cluster,color='orange')
plt.scatter(df77.user_id,df77.cluster,color='black')
plt.scatter(df88.user_id,df88.cluster,color='pink')
plt.title('clustring base on personality')
```
